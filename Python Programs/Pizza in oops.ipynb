```python
class Customer:
    def validate_quantity(self, quantity):
        if quantity >= 1 and quantity <= 5:
            return True
        else:
            return False
    
    def validate_pizza_type(self, pizza_type):
        if pizza_type.lower() == 'small' or pizza_type.lower() == 'medium':
            return True
        else:
            return False


class PizzaService:
    counter = 100
    def __init__(self, quantity, pizza_type, additional_topping):
        self.quantity = quantity
        self.pizza_type = pizza_type
        self.additional_topping = additional_topping
        self.pizza_cost = 0

    def calculate_pizza_cost(self):
        customer = Customer()
        if not customer.validate_quantity(self.quantity) or not customer.validate_pizza_type(self.pizza_type):
            self.pizza_cost = -1
            return

        pizza_cost_table = {
            'small': 150,
            'medium': 200
        }

        pizza_cost = pizza_cost_table[self.pizza_type.lower()]
        if self.additional_topping:
            pizza_cost += 35 if self.pizza_type.lower() == 'small' else 50

        self.pizza_cost = pizza_cost * self.quantity
        self.service_id = f'{self.pizza_type[0].upper()}{PizzaService.counter}'
        PizzaService.counter += 1


class DoorDelivery(PizzaService):
    def __init__(self, quantity, pizza_type, additional_topping, distance_in_kms):
        super().__init__(quantity, pizza_type, additional_topping)
        self.distance_in_kms = distance_in_kms

    def validate_distance_in_kms(self):
        if self.distance_in_kms >= 1 and self.distance_in_kms <= 10:
            return True
        else:
            return False

    def calculate_pizza_cost(self):
        if not self.validate_distance_in_kms():
            self.pizza_cost = -1
            return

        super().calculate_pizza_cost()

        if self.pizza_cost != -1:
            delivery_charge_table = {
                1: 5,
                2: 5,
                3: 5,
                4: 5,
                5: 5,
                6: 7,
                7: 7,
                8: 7,
                9: 7,
                10: 7
            }
            delivery_charge = delivery_charge_table[self.distance_in_kms]
            self.pizza_cost += delivery_charge * self.quantity

pizza1 = PizzaService(2, 'small', True)
pizza1.calculate_pizza_cost()
print(f'Service ID: {pizza1.service_id}, Pizza cost: {pizza1.pizza_cost}')

pizza2 = DoorDelivery(3, 'medium', False, 7)
pizza2.calculate_pizza_cost()
print(f"Service ID: {pizza2.service_id}, Pizza cost: {pizza2.pizza_cost}")
```
